## Representación gráfica de datos
Si la función de trazado se le da a una matriz de datos, la usará como coordenadas en el eje vertival, utilizará el índice de cada punto de datos en el Array  como la coordenada horizontal.

```python
plt.plot([1, 2, 3, 4, 5, 7, 8, 3, 1])
plt.show()
```

También se puede proporcionar dos matrices: una para el eje horizontal y otra para el eje horizontal

```python
plt.plot([1, 2, 3, 4, 5, 7, 8, 3, 1])
plt.show()
```

Pueden modificarse las longitudes de los ejes para que la figura no se vea tan ajustada

```python
plt.plot([-3, -1, 0, 4, 7],[1,4,6,7,6])
plt.show()
```

Se sigue el mismo procedimiento para pintar una función matemática

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y= (x**2)

plt.plot(x, y)
plt.show()
```

Tambien pude midificarse el estilo de la gráfica para que contenga más información.

```python
plt.plot(x, y)
plt.title("Función cuadrática")
plt.xlabel("X")
plt.ylabel("Y = x**2")
plt.grid(True);
plt.show()
```

Puuden superponerse gráficas y cambair el estilo de las lineas

```python
import numpy as np
x= np.linspace(-2,2,500)
y= x**2
y2 = x + 1
plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

Para poder diferenciar entre ambas funciones siempre es recomendado añadir una leyenda

```python
import numpy as np
x= np.linspace(-2,2,500)
y= x**2
y2 = x + 1
plt.plot(x, y, 'b--', label="x**2")
plt.plot(x, y2, 'g', label="x+1")
plt.show()
```

Para que las graficas no queden tan ajustadas, se puede hacer la figura mas grande

```python
plt.figure(figsize = (14, 6))
plt.subplot(1, 2, 1) #1 Rows, 2 Columns, 2st SUbplot
```

## Scatter plots

```python
from numpy.random import rand
x, y= rand(2,100)
plt.scatter(x,y)
plt.show()
```

```python
from numpy.random import rand
x, y = rand(2,100)
x2, y2= rand(2,100)
plt.scatter(x,y, c= 'red')
plt.scatter(x2, y2, c= 'blue')
plt.show()
```

## HIstogramas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3 ]
plt.subplot(211)# 2 rows, 1 column, 1st subplot
plt.hist(data, bins =10, rwidth = 0.8)

plt.subplot(212)# 2 rows, 1 column, 2nd subplot
plt.hist(data, bins =[1, 1.5, 2, 2.5, 3], rwidth = 0.95)
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

## Guardar la figura

```python
from numpy.random import rand
x, y = rand(2,100)
x2, y2= rand(2,100)

plt.scatter(x,y, c= 'red')
plt.scatter(x2, y2, c= 'blue')
plt.legend(loc = "best")
plt.savefig("Mi_Grafiquita.png", transparent = True)
plt.show()
```
